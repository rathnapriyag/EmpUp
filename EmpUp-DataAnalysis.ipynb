We will clean the Stage column by converting series x related values to series x (for example, pre series A will be rounded to series A). We do this so that we are able to work with the broader stages and not get bogged down with the finer details.

```python
import re 
com_st_df['Stage'] = com_st_df['Stage'].str.lower()
com_st_df['Stage'] = com_st_df['Stage'].apply(lambda x:  re.sub('-',' ',str(x) ))
com_st_df['Stage'] = com_st_df['Stage'].apply(lambda x:  re.sub('round|financing|funding|fund|investment|\+',' ',str(x) ))
com_st_df['Stage'] = com_st_df['Stage'].apply(lambda x:  re.sub('.*series.*(i|h|f|e|d|c|b|a).*',r'series \1',str(x)))
com_st_df['Stage'] = com_st_df['Stage'].apply(lambda x:  re.sub('.*seed.*','seed',str(x)))
com_st_df['Stage'] = com_st_df['Stage'].apply(lambda x:  re.sub('\$.*','nan',str(x)))
com_st_df['Stage'] = com_st_df['Stage'].apply(lambda x:  re.sub('$^','nan',str(x)))
com_st_df['Stage'] = com_st_df['Stage'].apply(lambda x:  re.sub('http.*','nan',str(x) ))

com_st_df['Stage'] = com_st_df['Stage'].replace({
    'seies a':'series a',
    'mid series':'pre series',
    'pe':'private equity'})

```

```python
# Converting multiple spaces into single space
com_st_df['Stage'] = com_st_df['Stage'].str.split().str.join(" ")
```

```python
# Converting to lowercase
com_st_df['Stage'] = com_st_df['Stage'].str.lower()
```

```python
# we will use the below function to remove the outliers
def reject_outliers(data, m = 2):
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else 0.
    return data[s<m]
```

```python
funds_amount_outlier_free  = reject_outliers(np.array(funds_final_date['amount'].dropna()))
funds_amount_outlier_free
```

```python
# Function for creating a vector from the input sentence
def get_sentence_embeding(sentences):
    preprocessed_text = bert_preprocess(sentences)
    return bert_encoder(preprocessed_text)['pooled_output']
```
